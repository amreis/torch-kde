# Differentiable Tophat Kernel Approximation

In this notebook, we demonstrate a differentiable approximation of the tophat kernel using a generalized Gaussian.

```python
# load required packages
import torch
import matplotlib.pyplot as plt
import scienceplots

from torchkde.kernels import TopHatKernel 
```

```python
BETA_GRID = [1., 2., 4., 8., 16.]
BW = 1.
```

```python
# generate grid
x = torch.linspace(-5, 5, 1000).unsqueeze(1)

plt.style.use('science')
plt.rcParams["figure.figsize"] = (6, 4)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Computer Modern Roman']
plt.xlabel('$x$')
plt.ylabel('Density')
plt.title('Differentiable Tophat Kernel Approximations')
for beta in BETA_GRID:
    kernel = TopHatKernel(beta)
    kernel.bandwidth = BW
    y = kernel(x)
    plt.plot(x, y, label=fr'$\beta = {beta}$')
plt.legend()
plt.savefig('plots/tophat_approx.pdf')
plt.savefig('plots/tophat_approx.svg')
plt.show()
```
